#Connecting to the db file

```python
db =sql.connect("chinook.db")
cursor = db.cursor()
```

```python
# showing all tables in the db
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
#showing the number of rows in a particular table
cursor.execute("SELECT COUNT (*) FROM media_types;")
print(cursor.fetchone())
```

```python
# printing data in a table
cursor.execute("SELECT * FROM albums LIMIT 10;")
print(cursor.fetchall())
```

```python
# printing table info
cursor.execute("PRAGMA table_info(employees);")
print(cursor.fetchall())
```

```python
cursor.execute("PRAGMA table_info(albums);")
print(cursor.fetchall())
```

```python
# connecting sql db to pandas df
df = pd.read_sql("SELECT CustomerId, FirstName, LastName, Country FROM customers;", db, index_col= 'CustomerId')
df.head()
```

```python
query = "SELECT * FROM artists;"
df = pd.read_sql(query, db, index_col= "ArtistId")
df.head()
```

```python
# printing table elements in order
query = "SELECT CustomerId, FirstName, LastName, Country FROM customers ORDER BY FirstName;"
df = pd.read_sql(query, db, index_col= "CustomerId")
df.head()
```

```python
# showing distinct countries
query = "SELECT DISTINCT Country FROM customers;"
df = pd.read_sql(query, db)
df
```

```python
# creating filter
query = "SELECT * FROM customers WHERE Country = 'United Kingdom';"
df = pd.read_sql(query, db, index_col= 'CustomerId')
df
```

```python
# Closing the db
db.close()
```
