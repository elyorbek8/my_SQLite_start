# Converting Pandas DataFrame to SQL database

```python
import pandas as pd
import sqlite3 as sql
```

```python
#creating a new empty db
db = sql.connect("newdb.db")
cursor = db.cursor()
```

```python
# uploading a df
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/Electricity_Production_By_Source.csv")
df.head()
```

```python
# processing df
df.drop(columns = ["Code"], inplace = True)
```

```python
# shortening column names
df.columns = ['country', 'year', 'coal', 'gas', 'hydro', 'renewables', 'solar', 'oil', 'wind', 'nueclear']
df
```

```python
# converting df to db
df.to_sql("electricity", db, if_exists='replace')
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM electricity Limit 10;")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM electricity WHERE country = 'Uzbekistan';", db)
```

#Creating a db table

```python
# creating a table
query = """
CREATE TABLE IF NOT EXISTS animals (
  id INT PRIMARY KEY,
  name VARCHAR(100),
  age INT,
  habitat VARCHAR(100)
);
"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
cursor.execute("PRAGMA table_info(animals)")
cursor.fetchall()
```

```python
# inserting values into a table
query = """
INSERT INTO animals (id, name, age, habitat)
VALUES
(1, 'Graffe', 7, 'Savanna'),
(2, 'Pengiun', 3, 'Antarctica'),
(3, 'Wolf', 4, 'Mountain');
"""

cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM animals;")
cursor.fetchall()
```

```python
df = pd.read_sql("SELECT * FROM animals;", db, index_col= 'id')
df
```

```python
cursor.execute("INSERT INTO animals (id, name) VALUES (4, 'Tiger');")
db.commit()
```

```python
# updating values
cursor.execute("UPDATE animals SET age = 6, habitat = 'Jungle' WHERE id = 4")
db.commit()
```

```python
cursor.execute("SELECT * FROM animals;")
cursor.fetchall()
```

```python
# deleting a particular value
cursor.execute("DELETE FROM animals WHERE name =  'Wolf';")
db.commit()
```

```python
cursor.execute("SELECT * FROM animals;")
cursor.fetchall()
```

```python
query = """
CREATE TABLE IF NOT EXISTS students (
  id INT PRIMARY KEY,
  name VARCHAR(100),
  age INT,
  major VARCHAR(100)
);
"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type = 'table';")
cursor.fetchall()
```

```python
# deleting a table
cursor.execute("DROP TABLE students;")
db.commit()
```
