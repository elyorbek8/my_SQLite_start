# Filtering for a particular country data

```python
import pandas as pd
import sqlite3 as sql
```

```python
# connecting a db
db = sql.connect('electricity.db')
cursor = db.cursor()
```

```python
# printing all tables in the db
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
# showing values in the table
cursor.execute("SELECT * FROM elect LIMIT 15;")
cursor.fetchall()
```

```python
# converting db to df
df = pd.read_sql("SELECT * FROM elect;", db)
df.head(15)
```

```python
# separating distinct countries
countries = pd.read_sql("SELECT DISTINCT country FROM elect;", db)
countries.head()
```

```python
# checking if a particular country is in the table
'Uzbekistan' in countries['country'].values
```

```python
# creating a new df for the country
dfuz = pd.read_sql("SELECT * FROM elect WHERE country = 'Uzbekistan';", db)
dfuz
```

```python
# deleting unnecessary columns
dfuz.drop(columns=['index', 'code', 'renewable_other', 'solar', 'wind', 'nuclear'], inplace= True)
```

# Some advanced SQL filtering operators: BETWEEN, IN, LIKE

```python
# filtering
query = "SELECT country, year, coal, gas, oil, hydro FROM elect WHERE country = 'Uzbekistan' AND year > 2015; "
pd.read_sql(query, db)
```

```python
query = "SELECT country, year, coal, gas, oil, hydro FROM elect WHERE country = 'Uzbekistan' AND year BETWEEN 2003 AND 2012; "
pd.read_sql(query, db)
```

```python
query = "SELECT country, year, coal, gas, oil, hydro FROM elect WHERE country = 'Uzbekistan' AND year IN (2003, 2005, 2008, 2015); "
pd.read_sql(query, db)
```

```python
query = "SELECT country, year, coal, gas, oil, hydro FROM elect WHERE country = 'Uzbekistan' AND year > 2015; "
pd.read_sql(query, db)
```

```python
# LIKE operator
query = "SELECT * FROM elect WHERE country LIKE '%ain' AND year = 2010;"
pd.read_sql(query, db)
```
